```python
import random
import uuid
from datetime import datetime, timedelta, time
from pathlib import Path
import numpy as np
import pandas as pd
import osmnx as ox

# ======================================================
# LOAD GRAPH
# ======================================================
print("🗺️ Loading wards graph...")
G = ox.load_graphml("hcmc_wards_drive.graphml")
nodes_gdf, _ = ox.graph_to_gdfs(G)

# ======================================================
# TIME SETTINGS
# ======================================================
WORK_START = 7
WORK_END = 23

SERVICE_DEADLINES = {
    "3h": 20,
    "5h": 18,
    "ban_tai_nhanh": 20,
    "ban_tai_4h": 19,
    "luu_kho": 22   # IMPORTANT: no longer unlimited
}

# ======================================================
# DEMAND SETTINGS
# ======================================================
random.seed(42)
np.random.seed(42)

START_DATE = datetime(2025, 5, 7)
SIM_DAYS = 220

BASE_DAILY_ORDERS = 5.7  # KEEP (now works correctly)

BIN_MINUTES = 10
BIN_SECONDS = BIN_MINUTES * 60

SCALE_FACTORS = {"moderate": 8, "nominal": 13, "stress": 15}

# ======================================================
# HOTSPOTS
# ======================================================
print("🎯 Creating demand hotspots...")
HOTSPOT_CENTERS = np.random.choice(list(G.nodes), size=40, replace=False)
HOTSPOT_PROB = 0.9

# ======================================================
# TEMPORAL PROFILE (NHPP)
# ======================================================
ci = np.array([
    2,1,3,1,6,5,7,8,6,7,13,15,13,23,13,41,38,64,73,79,82,97,89,
    80,94,88,68,50,74,48,55,47,35,39,32,33,33,33,32,34,29,27,32,26,11,20,22,18,16,21,19,14,21,29,24,22,15,13,14,14,16,
    8,17,10,18,9,3,6,10,7,11,3,6,3,2,4,13,4,7,2,3,3,5,10,11,9,16,7,6,6,5,4,3,13,8,4,13
])

pi = ci / ci.sum()
NUM_BINS = len(pi)
bin_offsets = np.arange(NUM_BINS) * BIN_SECONDS

# ======================================================
# ✅ FIX: FILTER ONLY WORKING HOURS + RENORMALIZE
# ======================================================
valid_bins = []
valid_pi = []

for b in range(NUM_BINS):
    hour = (bin_offsets[b] / 3600) % 24
    if WORK_START <= hour <= WORK_END:
        valid_bins.append(b)
        valid_pi.append(pi[b])

valid_pi = np.array(valid_pi)
valid_pi = valid_pi / valid_pi.sum()  # CRITICAL

# ======================================================
# SERVICE SETTINGS
# ======================================================
SERVICE_TYPES = {
    "3h": 3,
    "5h": 5,
    "ban_tai_nhanh": 2,
    "ban_tai_4h": 4,
    "luu_kho": 24
}

# ======================================================
# DISTANCE
# ======================================================
def haversine_distance(lat1, lon1, lat2, lon2):
    R = 6371000
    phi1, phi2 = np.radians([lat1, lat2])
    dphi = np.radians(lat2 - lat1)
    dlambda = np.radians(lon2 - lon1)
    a = np.sin(dphi / 2)**2 + np.cos(phi1)*np.cos(phi2)*np.sin(dlambda/2)**2
    return 2 * R * np.arctan2(np.sqrt(a), np.sqrt(1 - a))

# ======================================================
# LOCATION SAMPLING
# ======================================================
def random_location():
    if random.random() < HOTSPOT_PROB:
        node = random.choice(HOTSPOT_CENTERS)
    else:
        node = random.choice(list(G.nodes))
    return G.nodes[node]['y'], G.nodes[node]['x']

def generate_valid_pair():
    for _ in range(200):
        s_lat, s_lng = random_location()
        r_lat, r_lng = random_location()
        d = haversine_distance(s_lat, s_lng, r_lat, r_lng)
        if 1000 <= d <= 11000:
            return s_lat, s_lng, r_lat, r_lng
    return s_lat, s_lng, r_lat, r_lng

# ======================================================
# ✅ FIXED SERVICE SAMPLING
# ======================================================
def sample_service(created_hour):
    service_probs = {
        "3h": 0.6757,
        "5h": 0.3013,
        "ban_tai_4h": 0.01209,
        "ban_tai_nhanh": 0.0058,
        "luu_kho": 0.00511
    }

    candidates = []
    probs = []

    for s, p in service_probs.items():
        if created_hour <= SERVICE_DEADLINES[s]:
            candidates.append(s)
            probs.append(p)

    if not candidates:
        return "3h"

    probs = np.array(probs) / np.sum(probs)
    return np.random.choice(candidates, p=probs)

# ======================================================
# DELIVERY TIME
# ======================================================
def calculate_delivery_time(created, service):
    if service == "luu_kho":
        next_day = created.date() + timedelta(days=1)
        return datetime.combine(next_day, time(11, 0))
    return created + timedelta(hours=SERVICE_TYPES[service])

# ======================================================
# MAIN GENERATOR
# ======================================================
def generate_orders():
    rows = []

    for day in range(SIM_DAYS):
        scenario = random.choices(
            list(SCALE_FACTORS.keys()),
            weights=[0.33, 0.55, 0.12]
        )[0]

        scale = SCALE_FACTORS[scenario]
        mu_day = BASE_DAILY_ORDERS * scale

        day_start = START_DATE + timedelta(days=day)

        for i, b in enumerate(valid_bins):
            expected = mu_day * valid_pi[i]
            n = np.random.poisson(expected)

            for _ in range(n):
                created = day_start + timedelta(
                    seconds=bin_offsets[b] + np.random.uniform(0, BIN_SECONDS)
                )

                created_hour = created.hour + created.minute / 60

                service = sample_service(created_hour)
                s_lat, s_lng, r_lat, r_lng = generate_valid_pair()

                rows.append({
                    "id": str(uuid.uuid4()),
                    "scenario": scenario,
                    "createdAt": created,
                    "expectedDeliveryTime": calculate_delivery_time(created, service),
                    "senderLat": s_lat,
                    "senderLng": s_lng,
                    "receiverLat": r_lat,
                    "receiverLng": r_lng,
                    "serviceType": service,
                    "distance_meters": haversine_distance(s_lat, s_lng, r_lat, r_lng)
                })

    df = pd.DataFrame(rows).sort_values("createdAt").reset_index(drop=True)

    print(f"\n📦 TOTAL ORDERS: {len(df):,}")
    print(df["serviceType"].value_counts(normalize=True) * 100)

    return df

# ======================================================
# RUN
# ======================================================
print("🚀 Generating FIXED orders...")
df = generate_orders()

output = Path.home() / "Downloads" / "synthetic_orders_FIXED.csv"
df.to_csv(output, index=False)

print(f"\n✅ Saved to: {output}")
```

```python
import random
import math

EARTH_KM_PER_DEG = 111


def random_point_in_zone(lat, lon, radius_km):

    r = radius_km * math.sqrt(random.random())
    theta = random.uniform(0, 2 * math.pi)

    dx = r * math.cos(theta)
    dy = r * math.sin(theta)

    new_lat = lat + dy / EARTH_KM_PER_DEG
    new_lon = lon + dx / (EARTH_KM_PER_DEG * math.cos(math.radians(lat)))

    return new_lat, new_lon
```
