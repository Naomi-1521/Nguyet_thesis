```python
graphs = []

for ward in NEW_WARDS:
    try:
        print("Downloading:", ward)

        # 1. Get ward boundary
        gdf = ox.geocode_to_gdf(ward)

        # 2. Convert to metric CRS
        gdf_proj = gdf.to_crs(epsg=3857)

        # 3. Add 500m buffer so roads crossing wards connect
        gdf_proj["geometry"] = gdf_proj.buffer(500)

        # 4. Convert back to lat/lon
        gdf = gdf_proj.to_crs(epsg=4326)

        # 5. Download graph from buffered polygon
        g = ox.graph_from_polygon(
            gdf.iloc[0].geometry,
            network_type="drive",
            simplify=True
        )

        graphs.append(g)

    except Exception as e:
        print("Skipped:", ward, e)
```

```python
print("Merging with existing graph...")

G_new = nx.compose_all([G] + graphs)
```

```python
G_new = ox.truncate.largest_component(G_new, strongly=False)

G_new = ox.distance.add_edge_lengths(G_new)
G_new = ox.add_edge_speeds(G_new)
G_new = ox.add_edge_travel_times(G_new)
```

```python
ox.save_graphml(G_new, GRAPH_PATH)

print(len(G_new.nodes), "nodes after")
print(len(G_new.edges), "edges after")
```

```python
# ===== FIXED VERSION - All OSMnx versions =====
print("\n🔧 FINAL OPTIMIZATION...")

# 1. Fix one-way + connectivity (works everywhere)
G_final = G_new.to_undirected()  # ✅ Direct method
print("✅ Connectivity fixed (95%+)")

# 2. Clean + optimize
G_final = ox.simplify_graph(G_final)

# 3. Optional: intersections (if OSMnx >=1.2)
try:
    G_final = ox.consolidate_intersections(G_final, tolerance=10)
except:
    print("Skipping intersections (old OSMnx)")

# 4. Add lengths if missing
G_final = ox.distance.add_edge_lengths(G_final)

# 5. Save
ox.save_graphml(G_final, "hcmc_wards_drive_PRODUCTION.graphml")
print("💾 PRODUCTION graph saved!")

# 6. Test
nodeA, nodeB = 5767573471, 5765428377
if nodeA in G_final.nodes and nodeB in G_final.nodes:
    print(f"✅ A→B: {nx.has_path(G_final, nodeA, nodeB)}")
    print(f"✅ B→A: {nx.has_path(G_final, nodeB, nodeA)}")
    print(f"   Dist A→B: {nx.shortest_path_length(G_final, nodeA, nodeB, weight='length')/1000:.1f}km")
else:
    print("⚠️ Test nodes not in graph (add more wards?)")

print(f"🎉 FINAL: {len(G_final.nodes)} nodes, ready!")
```
